```python
from splinter import Browser
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
from pprint import pprint
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#Music Lessons Search on Groupon
url = 'https://www.groupon.com/browse/central-jersey?lat=40.058&lng=-74.406&administrative_area=NJ&address=New+Jersey&division=central-jersey&locale=en_US&sort=distance&query=music+lessons'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
html
#Find Activity Title
titles = soup.find_all('div', class_='cui-udc-details cui-udc-left-aligned-details')
titles
```

```python
Bootsdata={}
for i in range(len(titles)):
#     = titles[i].find('a')
#    title=l_0['title']
#    #listings[0]
   listings1= titles[i].find("div", class_="cui-udc-title one-line-truncate")
   name=listings1.get_text()
   Bootsdata.append({'name':name})
#                     'price':price})
Bootsdata
```
